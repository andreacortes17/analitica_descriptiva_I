```python
# Cargar las librerías necesarias para lectura de datasets
import pandas as pd
import os
# Cargar librerias de visualizacion de datos
import matplotlib.pyplot as plt
import numpy as np


# Cargar los datos en un DataFrame
dataset = pd.read_csv('/content/dataset_sintetico.csv', sep = ',')
```

```python
# Mostrar las primeras filas del DataFrame
print(dataset.head(2))
print("="*100)

# Mostrar la estructura del DataFrame
print(dataset.info())
print("="*100)

# Mostrar un resumen estadístico de las variables numéricas
print(dataset.describe())
print("="*100)
```

Con el head comprendemos la distribucion de nuestros datos y concluimos que es una tabla de un historial crediticio o base de datos de algun banco.
Con info podemos separa nuestros datos en los siguientes tipos:<br>
**ID:** numerico, discreto<br>
**Edad:** cuantitativo discreto<br>
**Salario:** cuantitativo continuo<br>
**Puntuacion de credito:** cuantitativo continuo<br>
**Genero:** cualitativo categorico<br>
**Estado civil:** cualitativo categorico<br>
**Compras anuales:** cuantitativo discreto<br>
**Deuda:** cuantitativo continuo<br>

```python
# Seleccionar solo las columnas numéricas
columnas_numericas = dataset.select_dtypes(include=[np.number]).columns

# Excluir la columna de ID
columnas_numericas = [col for col in columnas_numericas if col != 'ID']

# Calcular el número de filas y columnas para el subplot
n = len(columnas_numericas)
nrows = 3
ncols = min(n, 3)

# Crear la figura y los subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 5*nrows))
fig.suptitle('Distribución de Variables Numéricas', fontsize=16)

# Aplanar el array de ejes en caso de que sea 2D
axes = axes.flatten() if n > 3 else [axes]

# Crear histogramas para cada variable numérica
for i, col in enumerate(columnas_numericas):
    ax = axes[i]
    dataset[col].hist(ax=ax, bins=50, edgecolor='black')
    ax.set_title(f'Distribución de {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frecuencia')

# Ocultar subplots vacíos si los hay
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Podemos identificar la distribucion de las diferentes edades, salarios, p. de credito, cuanto compran anualmente y cuanta deuda poseen las personas registradas en la base de datos del campo.
En las edades podemos ver que hay tendencias a las personas mas jovenes entre 18 a 20 seguramente, segun nuestro grafico, y mayores de 55.
el salario es mas equitativo pero tiene tambien a repetirse mas los valores mas pequeños y los sueldos intermedios.
En p. credito podemos ver que la mayoria tiene puntuaciones de credito bajas.
Otro dato interesante seria la distribucion del credito ya que nos da a entender que la mayoria posee deudas con el banco por cantidades "pequeñas", con un descenso despues de los 10 mil.
